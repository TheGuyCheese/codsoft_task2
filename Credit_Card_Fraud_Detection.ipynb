```python
X = data_train.iloc[:, :21]
Y = data_train.iloc[:, 22]
X = X.drop(columns=['Unnamed: 0', 'trans_date_trans_time', 'merchant', 'first', 'last', 'street', 'city', 'state', 'zip', 'job', 'dob', 'trans_num','category','gender'])
print(X.shape)
print(Y.shape)
```

```python
X.head()
```

```python
X, Y = make_classification(random_state=42)
X_train, X_test, y_train, y_test=train_test_split(X, Y, test_size=0.2)
baseline_model1 = Pipeline([
    ("Standard_Scaler", StandardScaler()),
    ("Logistic_Regression", LogisticRegression())
])
baseline_model1.fit(X_train, y_train)
```

```python
baseline_score1 = baseline_model1.score(X_test, y_test)
baseline_score1
```

```python
from sklearn.tree import DecisionTreeClassifier

dts_cls = DecisionTreeClassifier(random_state=42)

baseline_model2 = Pipeline([
    ("Standard_Scaler", StandardScaler()),
    ("Descision_Tree", dts_cls)
])

baseline_model2.fit(X_train, y_train)
```

```python
baseline_score2 = baseline_model2.score(X_test, y_test)
baseline_score2
```

```python
from sklearn.ensemble import RandomForestClassifier

ran_cls = RandomForestClassifier(n_estimators=100, random_state=42)

baseline_model3 = Pipeline([
    ("Standard_Scaler", StandardScaler()),
    ("Descision_Tree", ran_cls)
])

baseline_model3.fit(X_train, y_train)
```

```python
baseline_score3 = baseline_model3.score(X_test, y_test)
baseline_score3
```

```python
accuracies = {'Logistic Regression' : baseline_score1,
              'Decision Tree' : baseline_score2,
              'Random Forest' : baseline_score3}
df = pd.DataFrame.from_dict(accuracies, orient='index')
df.head()
```
